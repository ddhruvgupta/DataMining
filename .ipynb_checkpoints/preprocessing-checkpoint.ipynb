# Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./bank/bank.csv", sep=";")
```

```python
df.head()
```

```python
df.tail()
```

```python
def replace_marital(val):
    if val == "single":
        return 0
    else:
        return 1

df["marital"] = df["marital"].apply(replace_marital, 1)
```

```python
df.head()
```

```python
df["housing"] = df["housing"].map({
    "no": 0,
    "yes": 1
}.get)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["loan"] = df["loan"].replace({
    "no": 0,
    "yes": 1
})
```

```python
df.head()
```

```python
df.tail()
```

```python
df["job"].unique()
```

```python
df["job"].replace({
    'unknown': np.nan,
    'management': 0,
    'technician': 1,
    'entrepreneur': 2,
    'blue-collar': 3,
    'retired': 4,
    'admin.': 5,
    'services': 6,
    'self-employed': 7,
    'unemployed': 8,
    'housemaid': 9,
    'student': 10
}, inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["education"].unique()
```

```python
df["education"].replace({
    'unknown': np.nan,
    'tertiary': 0, 
    'secondary': 1, 
    'primary': 2
}, inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["default"].replace({
    "no": 0,
    "yes": 1
}, inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

<h2>Min-Max Normalization</h2>
<p><img src="https://qph.fs.quoracdn.net/main-qimg-0d692d88876aeb26b1f1a578d1c5a94e" /></p>

```python
df["balance"].min()
```

```python
df["balance"].max()
```

```python
saved_balance = df.balance.values
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(np.arange(len(saved_balance)), saved_balance, "r")

plt.show()
```

```python
df["balance"] = df["balance"].apply(lambda v: (v - df["balance"].min()) / (df["balance"].max() - df["balance"].min()))
```

```python
df.head()
```

```python
df.tail()
```

```python
saved_balance
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots(1, 2)
axes[0].plot(np.arange(len(saved_balance)), saved_balance, "r")
axes[1].plot(np.arange(len(df.balance)), df.balance, "g")
plt.show()
```

```python
df.contact.replace({"unknown": np.nan,
            "telephone": 0,
            "cellular": 1}, inplace=True)
```

```python
df.month = df.month.map({
    'oct': 10, 'may': 5, 'apr': 4, 'jun': 6, 'feb': 2, 'aug': 8, 'jan': 1, 'jul': 7, 'nov': 11,
       'sep': 9, 'mar': 3, 'dec': 12
})
```

```python
df.head()
```

```python
df.poutcome.unique()
```

```python
df.poutcome = df.poutcome.map({
    'unknown': np.nan, 
    'failure': 0, 
    'other': 1, 
    'success': 2
})
```

```python
df.head()
```

```python
df.pdays = df.pdays.apply(lambda v: (v-df.pdays.min())/(df.pdays.max()-df.pdays.min()))
```

```python
df.head()
```

```python
df.y.replace({"no": 0, "yes": 1}, inplace=True)
```

```python
df.duration = df.duration.apply(lambda v: ((v-df.duration.min())/(df.duration.max()-df.duration.min())))
```

    <h1>Outliers</h1>
<p>Now, how do we deal with outliers? Here are four approaches:</p>
<p><b>1.&nbsp;</b><b>Drop the outlier records.</b></p>
<p>In the case of Bill Gates, or another true outlier, sometimes it’s best to completely remove that record from your dataset to keep that person or event from skewing your analysis.</p>
<p><b>2. Cap your data.</b></p>
<p>Another way to handle true outliers is to cap them. For example, if you’re using income, you might find that people above a certain income level behave in the same way as those with a lower income. In this case, you can cap the income value at a level that keeps that intact.</p>
<p><b>3. Assign a new value.</b></p>
<p>If an outlier seems to be due to a mistake in your data, you try imputing a value. Common <a href="http://en.wikipedia.org/wiki/Imputation_(statistics)">imputation methods</a> include using the mean of a variable, or utilizing a regression model to predict the missing value.</p>
<p><b>4. Try a transformation.</b></p>
<p>A different approach to true outliers could be to try creating a transformation of the data rather than using the data itself. For example, try creating a percentile version of your original field and working with that new field instead.</p>
<p>Just how much an outlier affects your analysis depends, not surprisingly, on a few factors. One factor is dataset size. In a large dataset, each individual point carries less weight, so an outlier is less worrisome than the same data point would be in a smaller dataset. Another consideration is “how much” of an outlier a point might be – just how far out of line with the rest of your dataset a single point is. A point that is ten times as large as your upper boundary will do more damage than a point that is twice as large.</p>
